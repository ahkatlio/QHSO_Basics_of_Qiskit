    <h1 style="text-align: center;">The Bloch Sphere: Your Quantum Compass</h1>
    <div class="center">
        <img src="Bloch_Sphere_animations/qiskit.gif" alt="Bloch sphere_animation" width="200">
    </div>
    <p>Imagine you have a magical globe, but instead of representing Earth's geography, it represents the possible states of a qubit. This is the Bloch sphere. It's a way to visualize and understand the unique characteristics of quantum bits or qubits.</p>
    <h2>Representing Quantum States</h2>
    <p>In classical computing, bits can be in one of two states: 0 or 1, much like flipping a coin with heads or tails. But qubits can exist in a whole range of states in between, and the Bloch sphere helps us grasp this concept.</p>
    <h2>The North Pole and South Pole</h2>
    <br>
    <div class="center">
        <img src="Bloch_Sphere_animations/bloch.png" alt="Bloch sphere" width="500">
    </div>
    <br>
    <p>At the top of the Bloch sphere is the "North Pole", representing the qubit state |0⟩ (like a classical bit being 0). At the bottom is the "South Pole", representing the qubit state |1⟩ (like a classical bit being 1). These are the two most straightforward states.</p>
    <h2>Superposition: The Magic of Quantum States</h2>
    <p>Where the Bloch sphere gets truly fascinating is the space in between the North and South Poles. This represents superposition, a state where a qubit can be both |0⟩ and |1⟩ simultaneously, like a spinning coin showing both heads and tails at once.</p>
    <h2>Rotations on the Sphere</h2>
    <p>Quantum operations, like gates (such as the Hadamard gate or Pauli gates), are represented as rotations on the Bloch sphere. These operations move the qubit's state around the sphere, changing its superposition and phase.</p>
    <h2>Measurement</h2>
    <p>When we measure a qubit, it "collapses" to one of the poles (North or South) based on the probabilities associated with its superposition. It's like the spinning coin landing either heads or tails when observed.</p>
    <h2>Quantum Entanglement</h2>
    <p>The Bloch sphere also helps us understand quantum entanglement, where the states of two qubits become intertwined. In the Bloch sphere representation, entanglement can be visualized as the qubits' states being correlated in a way that's beyond classical explanation.</p>
    <h2>Quantum Computing's Compass</h2>
    <p>In essence, the Bloch sphere is like a compass for quantum computing. It lets us navigate and understand the quantum states of qubits, making complex quantum operations more manageable to grasp.</p>

1) **Import Libraries**: We begin by importing the required functions and libraries from Qiskit. We import `plot_bloch_multivector` for Bloch sphere visualization, `QuantumCircuit` to create a quantum circuit, `Aer` to access simulators, and `execute` to run the circuit on a simulator.

```python
# Import the necessary functions from Qiskit
from qiskit.visualization import plot_bloch_multivector
from qiskit import QuantumCircuit, Aer, execute
```

2) **Create a Quantum Circuit**: We create a quantum circuit `qc` with one qubit. This circuit doesn't have any gates or operations yet; it's a blank slate.

```python
# Create a quantum circuit with one qubit
qc = QuantumCircuit(1)
```

3) **Choose a Simulator Backend**: We select the `statevector_simulator` backend from `Aer`. This backend allows us to simulate quantum circuits and obtain the quantum state (statevector) of the qubits.

```python
# Execute the quantum circuit on a statevector simulator backend
backend = Aer.get_backend('statevector_simulator')
```

4) **Execute the Quantum Circuit**: We use the `execute` function to run the quantum circuit (`qc`) on the selected backend (`backend`). The `result` variable stores the outcome of the simulation.

```python
# This code executes a quantum circuit `qc` on a backend and stores the result in the `result` variable.
# The `execute` function is used to run the circuit on the specified backend and returns a job object.
# The `result()` method is then called on the job object to get the result of the execution.
result = execute(qc, backend).result()
```

5) **Retrieve the Statevector**: We extract the quantum statevector of the qubits after executing the circuit. This statevector represents the quantum state of the qubit(s) in the circuit.

```python
# Get the statevector of the quantum circuit
statevector = result.get_statevector(qc)
```

6) **Visualize on the Bloch Sphere**: Finally, we use `plot_bloch_multivector` to visualize the obtained quantum state on the Bloch sphere. The Bloch sphere provides a graphical representation of the qubit's state, showing its direction and superposition if applicable.

```python
# Plot the statevector on the Bloch sphere
plot_bloch_multivector(statevector)
```
