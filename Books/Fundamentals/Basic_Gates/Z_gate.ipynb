    <h2>What is the Z Gate?</h2>
    <p>Imagine you have a magical tool that can change the way a qubit thinks. That tool is the Z gate. When we apply the Z gate to a qubit, it doesn't flip it like the X gate. Instead, it tweaks the phase of the qubit's state.</p>
    <ul><li><strong>X-Gate (Bit Flip)</strong>: Flips qubits between '0' and '1.'</li><li><strong>Y-Gate (The Twist)</strong>: Takes us on a spin around the quantum realm. </li></ul>
    <h2>But what do we mean by a spin?</h2>
    <p>It rotates the input around the y-axis of the Bloch sphere by 180 degrees.</p>
