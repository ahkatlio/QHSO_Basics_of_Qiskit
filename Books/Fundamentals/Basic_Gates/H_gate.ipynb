    <h1 style="text-align: center;">The Hadamard Gate (H gate)</h1>
    <p>In the enchanting world of quantum computing, the Hadamard gate (often called the H gate) is like a magician's wand, performing a mesmerizing trick with quantum bits or "qubits". This gate is one of the fundamental tools that give quantum computers their superpowers. In this section, we'll break down the Hadamard gate in simple terms.</p>
    <h2>What is the Hadamard Gate?</h2>
    <p>Imagine you have a quantum bit, or "qubit". A regular computer bit can be either a 0 or a 1, like a light switch being off or on. But the H gate is a special tool that allows qubits to do something extraordinary: it puts them in a state of "superposition". But what's superposition? Imagine it as a bit like Schrödinger's cat being both alive and dead at the same time until you look. In quantum terms, it means that the qubit can be in multiple states simultaneously. It's like having a spinning coin that's both heads and tails at the same time.</p>
    <h2>How Does the H Gate Work?</h2>
    <p>Here's the magic of the H gate:</p>
    <ol>
        <li>If you start with a qubit in the state |0⟩ (like a light switch off), applying the H gate turns it into a blend of |0⟩ and |1⟩. It's as if the switch is flickering between on and off!</li>
        <br>
        <br>
            <div class="center">
                <img src="H_Gate_animations/Hgate0.gif" alt="Hgate0" width="600">
            </div>
        <br>
        <br>
        <li>If you begin with a qubit in the state |1⟩ (like a light switch on), guess what? The H gate does the same thing, making it a mix of |0⟩ and <strong>-|1⟩</strong>. Now, it's like the switch is winking on and off!</li>
        <br>
        <br>
            <div class="center">
                <img src="H_Gate_animations/Hgate1.gif" alt="Hgate1" width="600">
            </div>
        <br>
        <br>
    </ol>
    <p> The H gate is a 90° rotation around the X+Z axis on the Bloch sphere. It's a fundamental tool in quantum computing, and it's used in many quantum algorithms. It's also the key to understanding the Bell state, which is the basis of quantum entanglement.</p>
    <h2>How to use the H Gate</h2>
    <p>Let's see how the H gate works in Qiskit.</p>
</body>

</html>

1) We import the necessary packages and set up the program:

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_multivector
```

2) We create a quantum circuit with one qubit and one classical bit:

```python
qc = QuantumCircuit(1,1)
```

3) We add the H gate to our circuit:

```python
qc.h(0)
```

4) We see the result in the Bloch sphere:

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc, backend).result()
statevector = result.get_statevector(qc)
plot_bloch_multivector(statevector)
```

5) We measure the qubit:

```python
qc.measure(0,0)
```

6) Let's see the circuit:

```python
qc.draw(output='mpl')
```

7) We run the program on a simulator:

```python
simulator = Aer.get_backend('qasm_simulator')
```

8) We see the results:

```python
result = execute(qc, backend=simulator, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

As We can see, the qubit is in a superposition of |0⟩ and |1⟩. The result is either 0 or 1, each with a 50% probability. Which one will it be? We won't know until we measure it! That's the magic of the H gate. The output is basically known as  $ \ket{+} $ state.

    <h2>What if we apply H Gate twice?</h2>
    <p>Let's create a circuit again:</p>
</body>

</html>

1) We create a quantum circuit with one qubit and one classical bit angain and add the H gate twice to our circuit:

```python
qc2 = QuantumCircuit(1,1)
qc2.h(0)
qc2.h(0)
qc2.measure(0,0)
qc2.draw(output='mpl')
```

2) We see the results using a simulator:

```python
result = execute(qc2, backend=simulator, shots=1024).result()
counts = result.get_counts()
plot_histogram(counts)
```

3) We see the result in the Bloch sphere:

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc2, backend).result()
statevector = result.get_statevector(qc2)
plot_bloch_multivector(statevector)
```

Here we can see the output is 0. Applying H Gate twice on 0 gives us 0 again. This is because the H gate is its own inverse. In other words, applying the H gate twice is the same as doing nothing at all. This is true for all single-qubit quantum gates.

    <h2>What if we apply an H gate, then measure it, and apply an H gate again?</h2>
    <p>Let's perform a quantum experiment that involves applying an H (Hadamard) gate, measuring, and then applying another H gate. This experiment demonstrates the probabilistic and reversible nature of quantum operations. Follow along step by step:</p>
</body>

</html>

1) Apply H-Gate

```python
qc4 = QuantumCircuit(1,1)
qc4.h(0)
```

2) Measure the qubit

```python
qc4.measure(0,0)
```

3) Apply H-Gate again

```python
qc4.h(0)
```

4) Draw the circuit

```python
qc4.draw(output='mpl')
```

5) See the results using a bloch sphere:

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc4, backend).result()
statevector = result.get_statevector(qc4)
plot_bloch_multivector(statevector)
```

See, each time we are runnig the `step 5`, we are getting either $ \ket{+} $ or $ \ket{-} $ state. But we applied H gate twice. So, we should get the initial state of the qubit, right? But why it's returning $ \ket{+} $ or $ \ket{-} $ state with 50% probability? Good question!

    <h2>Why?</h2>
    <p>Let's see what happed there:</p>
    <h2>Applying Two H-Gates Without Measurement in Between</h2>
    <p>When you apply an H-gate twice in a row without any measurement in between, here's what happens:</p>
    <ol>
        <li>Start with a qubit in the state |0⟩</li>
        <li>Apply the first H-gate, which transforms it into |+⟩ </li>
        <li>Apply the second H-gate, which transforms it into |0⟩ again. So, applying two H-gates in a row without measurement restores the qubit to its original state ∣0⟩. This demonstrates the reversible nature of quantum gates.</li>
    </ol>
    <h2>Applying Two H-Gates With a Measurement in Between</h2>
    <p>When you apply an H-gate twice in a row with a measurement in between, here's what happens:</p>
    <ol>
        <li>Start with a qubit in the state |0⟩</li>
        <li>Apply the first H-gate, which transforms it into |+⟩ </li>
        <li>Measure the qubit. It now has a 50% chance of being in either state |0⟩ or |1⟩.</li>
        <li>Let's say the result is |1⟩. Apply the second H-gate, which transforms it into |−⟩ </li>
        <li>Measure the qubit. It now has a 50% chance of being in either state |0⟩ or |1⟩.</li>
        <li>Let's say the result is |0⟩. Apply the second H-gate, which transforms it into |+⟩</li>
    </ol>
    <p>So, the crucial distinction is that measurement in between the two H-gates introduces uncertainty in the qubit's state, whereas without measurement, the qubit returns to its original state deterministically. This difference is a fundamental aspect of quantum computation, where measurements can alter the outcomes of subsequent operations. <strong>We will discuss the mathematical aspect of this phenomenon in the linear algebra notebook.</strong></p>
