    <h1 style="text-align: center;">Quantum Rotation Gates: Turning Quantum Bits with Precision</h1>
    <p>In this notebook, we'll explore an exciting concept called "rotation gates." These gates are like magic wands that allow us to turn and twist quantum bits (qubits) in precise ways. Let's dive in!</p>
    <h2>Let's see how Rotation Gates work</h2>
    <p>But first we are going to create an empty bloch sphere.</p>
</body>

</html>

```python
# Import the necessary functions from Qiskit
from qiskit.visualization import plot_bloch_multivector
from qiskit import QuantumCircuit, Aer, execute
from math import pi

# Create a quantum circuit with one qubit
qc = QuantumCircuit(1)

# Execute the quantum circuit on a statevector simulator backend
backend = Aer.get_backend('statevector_simulator')
result = execute(qc, backend).result()

# Get the statevector of the quantum circuit
statevector = result.get_statevector(qc)

# Plot the statevector on the Bloch sphere
plot_bloch_multivector(statevector)
```

We can see the arrow is pointing up, which means the qubit is in the state |0⟩. Let's see what happens when we apply a rotation gate to this qubit. We want to rotate this arrow 90 degrees around the y-axis. We can do this by applying the Ry gate with a rotation angle of 90 degrees.

```python
qc.ry(pi/3, 0)
```

Let's see how the circuit looks like:

```python
qc.draw(output='mpl')
```

Lst's see how the bloch sphere looks like after applying the Ry gate:

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc, backend).result()
statevector = result.get_statevector(qc)
plot_bloch_multivector(statevector)
```

What if we want to rotate the qubit 60 degrees around the x-axis? We can do this by applying the Rx gate with a rotation angle of 60 degrees.

```python
qc1 = QuantumCircuit(1)
qc1.rx(pi/3, 0)
```

Let's see how the circuit looks like:

```python
qc1.draw(output='mpl')
```

Let's see how the bloch sphere looks like after applying the Rx gate:

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc1, backend).result()
statevector = result.get_statevector(qc1)
plot_bloch_multivector(statevector)
```

Now let's see what happens when we apply the Rz gate to the qubit. We want to rotate this arrow 90 degrees around the z-axis. We can do this by applying the Rz gate with a rotation angle of 90 degrees.

```python
qc2 = QuantumCircuit(1)
qc2.rz(pi/2, 0)
```

Let's see how the circuit looks like:

```python
qc2.draw(output='mpl')
```

Let's see how the bloch sphere looks like after applying the Rz gate:

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc2, backend).result()
statevector2 = result.get_statevector(qc2)
plot_bloch_multivector(statevector2)
```
